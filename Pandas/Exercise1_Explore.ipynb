### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. What is the number of observations in the dataset?

```python
# Solution 1
num_observations = chipo.shape[0]
print(num_observations)
```

### Step 6. What is the number of columns in the dataset?

```python
num_columns = chipo.shape[1]
print(num_columns)
```

### Step 7. Print the name of all the columns.

```python
print(chipo.columns.tolist())
```

### Step 8. How is the dataset indexed?

```python
print(chipo.index)
```

### Step 9. Which was the most-ordered item?

```python
grouped = chipo.groupby('item_name')['quantity'].sum()
most_ordered = grouped.sort_values(ascending=False).head(1)
print(most_ordered)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
quantity = most_ordered.iloc[0]
print(quantity)
```

### Step 11. What was the most ordered item in the choice_description column?

```python
most_choice = (
    chipo.groupby('choice_description', as_index=False)['quantity']
    .sum()
    .sort_values('quantity', ascending=False)
    .head(1)
)

print(most_choice)
```

### Step 12. How many items were orderd in total?

```python
total_items = chipo['quantity'].sum()
print(total_items)
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
print(chipo.item_price.dtype)
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.replace('$','')))
```

#### Step 13.c. Check the item price type

```python
print(chipo.item_price.dtype)
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print(f'Revenue was: ${revenue:.2f}')
```

### Step 15. How many orders were made in the period?

```python
num_orders = chipo['order_id'].nunique()
print(num_orders)
```

### Step 16. What is the average revenue amount per order?

```python
order_revenue = chipo.groupby('order_id').apply(
    lambda df: (df.quantity * df.item_price).sum()
)
avg_revenue = order_revenue.mean()
print(avg_revenue)
```

### Step 17. How many different items are sold?

```python
num_unique_items = chipo['item_name'].nunique()
print(num_unique_items)
```
